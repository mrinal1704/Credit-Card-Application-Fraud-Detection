```python
for col in d:
    print(d[col].value_counts())
```

```python
for col in d:
    print(col)
    x=d[col].unique()
    print(len(x))
```

```python
categoricalVar=d.columns.tolist()
categoricalVar.remove('record')
for col in categoricalVar:
    df = pd.DataFrame(d[col].value_counts().sort_values(ascending=False).head(20))
    df.plot(kind='bar',figsize=(12,6))
    plt.xlabel(col,fontsize=15)
    plt.ylabel('Frequency',fontsize=15)
    plt.yscale('log')
    plt.xticks(fontsize=12)
```
